```python
import pandas as pd
from pandasai import PandasAI

# Sample dataset
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']
}

# Create a DataFrame
df = pd.DataFrame(data)

# Initialize PandasAI with the DataFrame
pandas_ai = PandasAI(df)

# Define a function to interact with the dataset
def chat_with_dataset(query):
    response = pandas_ai.query(query)
    return response

# Example usage
if __name__ == "__main__":
    while True:
        user_input = input("Ask a question about the dataset (or type 'exit' to quit): ")
        if user_input.lower() == 'exit':
            break
        result = chat_with_dataset(user_input)
        print("Response:", result)
```

Log all chat questions and responses into a new table in the postgresSQL database 
Create a new table in the PostgreSQL database for storing the queries and responses.
Insert records into this new table each time a query is processed.


Database Table Creation: You need to execute the SQL command to create the chat_log table in your PostgreSQL database. This table includes columns for storing queries and responses, as well as a timestamp.

Log Chat Function: The log_chat function takes a query and its corresponding response, and inserts them into the chat_log table. It uses SQLAlchemyâ€™s text function to create a parameterized SQL query for inserting data.

Integrate Logging: The chat_with_dataset function now calls log_chat to log each query and its response to the database before returning the response to the user.

Interactive Loop: The loop allows you to ask questions about the dataset. Each question and answer pair is logged into the chat_log table.

```python
#table_creatio

CREATE TABLE chat_log (
    id SERIAL PRIMARY KEY,
    query TEXT NOT NULL,
    response TEXT NOT NULL,
    timestamp TIMESTAMPTZ DEFAULT CURRENT_TIMESTAMP
);
```

```python
import pandas as pd
from sqlalchemy import create_engine, text
from pandasai import PandasAI

# PostgreSQL connection details
db_user = 'your_username'
db_password = 'your_password'
db_host = 'your_host'
db_port = 'your_port'
db_name = 'your_database'

# Create SQLAlchemy engine
engine = create_engine(f'postgresql://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}')

# Query the table
query = """
SELECT time, date, vehicle_registration_number_services_doe
FROM your_table_name
"""

# Read the query result into a DataFrame
df = pd.read_sql(query, engine)

# Initialize PandasAI with the DataFrame
pandas_ai = PandasAI(df)

# Define a function to insert chat logs into the database
def log_chat(query, response):
    with engine.connect() as connection:
        insert_query = text("""
            INSERT INTO chat_log (query, response)
            VALUES (:query, :response)
        """)
        connection.execute(insert_query, {'query': query, 'response': response})

# Define a function to interact with the dataset and log the chat
def chat_with_dataset(query):
    response = pandas_ai.query(query)
    log_chat(query, response)
    return response

# Example usage
if __name__ == "__main__":
    while True:
        user_input = input("Ask a question about the dataset (or type 'exit' to quit): ")
        if user_input.lower() == 'exit':
            break
        result = chat_with_dataset(user_input)
        print("Response:", result)
```

Take query from csv file 

```python


# Create SQLAlchemy engine
engine = create_engine(f'postgresql://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}')

# Query the table
query = """
SELECT time, date, vehicle_registration_number_services_doe
FROM your_table_name
"""

# Read the query result into a DataFrame
df = pd.read_sql(query, engine)

# Initialize PandasAI with the DataFrame
pandas_ai = PandasAI(df)

# Define a function to insert chat logs into the database
def log_chat(query, response):
    with engine.connect() as connection:
        insert_query = text("""
            INSERT INTO chat_log (query, response)
            VALUES (:query, :response)
        """)
        connection.execute(insert_query, {'query': query, 'response': response})

# Define a function to process queries from a CSV file
def process_queries_from_csv(csv_file):
    # Read queries from the CSV file
    queries_df = pd.read_csv(csv_file)
    
    # Ensure the CSV has a 'query' column
    if 'query' not in queries_df.columns:
        raise ValueError("CSV file must contain a 'query' column")

    # Process each query
    for index, row in queries_df.iterrows():
        query = row['query']
        response = pandas_ai.query(query)
        log_chat(query, response)
        print(f"Processed query: {query}")
        print(f"Response: {response}")

# Example usage
if __name__ == "__main__":
    csv_file = 'queries.csv'  # Path to your CSV file with queries
    process_queries_from_csv(csv_file)
```
